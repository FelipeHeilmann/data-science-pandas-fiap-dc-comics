```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./dc-comics.csv")
```

```python
df = pd.read_csv('./dc-comics.csv')

# 1. Distribuição de Personagens por Sexo
sex_counts = df['SEX'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Personagens por Sexo')
plt.show()

# 2. Distribuição de Personagens por Status de Vida
alive_counts = df['ALIVE'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(alive_counts.index, alive_counts.values)
plt.title('Distribuição de Personagens por Status de Vida')
plt.xlabel('Status de Vida')
plt.ylabel('Número de Personagens')
plt.show()

# 3. Distribuição de Personagens por Cor de Olhos
eye_color_counts = df['EYE'].value_counts().head(10)
plt.figure(figsize=(10, 6))
eye_color_counts.plot(kind='barh')
plt.title('Distribuição de Personagens por Cor de Olhos (Top 10)')
plt.xlabel('Número de Personagens')
plt.ylabel('Cor de Olhos')
plt.show()

# 4. Distribuição de Personagens por Cor de Cabelo
hair_color_counts = df['HAIR'].value_counts().head(10)
plt.figure(figsize=(10, 6))
hair_color_counts.plot(kind='barh')
plt.title('Distribuição de Personagens por Cor de Cabelo (Top 10)')
plt.xlabel('Número de Personagens')
plt.ylabel('Cor de Cabelo')
plt.show()

# 5. Distribuição de Personagens por Alinhamento (ALIGN)
align_counts = df['ALIGN'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(align_counts, labels=align_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Personagens por Alinhamento')
plt.show()

# 6. Ano de Criação dos Personagens
year_counts = df['YEAR'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar')
plt.title('Ano de Criação dos Personagens')
plt.xlabel('Ano')
plt.ylabel('Número de Personagens')
plt.xticks(rotation=45)
plt.show()


# 7. Distribuição dos 10 Personagens mais Recorrentes
top_characters = df['name'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_characters.plot(kind='bar')
plt.title('Top 10 Personagens mais Recorrentes')
plt.xlabel('Personagem')
plt.ylabel('Número de Aparições')
plt.xticks(rotation=45)
plt.show()
```
